```python
data[data['amount_spent']>percentil_95]

new_labels = ['Preferred','VIP']
data['Customer_Type'] = pd.cut(data['amount_spent'], [0,19.8,67.5], labels=new_labels)
data.head()
#data['Customer_Type'] = data['Customer_Type'].cat.add_categories('Normal')
data['Customer_Type'] = data['Customer_Type'].astype('category')
data['Customer_Type'] = data['Customer_Type'].cat.add_categories('Normal')
data['Customer_Type']=data['Customer_Type'].fillna('Normal')

data['Customer_Type'].unique()
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

data.head(10)

#VIP_customers =  data.groupby(['Country','Customer_Type'])['amount_spent'].agg(sum)


filter_VIP = data[data['Customer_Type']=='VIP']
#VIP_customers =  filter_VIP.groupby(['Country','Customer_Type'])['amount_spent'].agg(sum)
filter_VIP

#VIP_country = filter_VIP.groupby(['Country'])['CustomerID'].agg(count)

grouped = filter_VIP.groupby('Country').agg({'CustomerID': 'count'}).reset_index()
grouped
grouped_ordered = grouped.sort_values(by='CustomerID', ascending=False) #los ordeno de mayor a menor
grouped_ordered.head(10) 
print(grouped_ordered.iloc[:5]['Country'])
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here

filter_VIP_Preferred  = data[data['Customer_Type']!="Normal"]
#VIP_customers =  filter_VIP.groupby(['Country','Customer_Type'])['amount_spent'].agg(sum)
filter_VIP_Preferred

#VIP_country = filter_VIP.groupby(['Country'])['CustomerID'].agg(count)

grouped_2 = filter_VIP_Preferred.groupby('Country').agg({'CustomerID': 'count'}).reset_index()
grouped_2
grouped_ordered_2 = grouped_2.sort_values(by='CustomerID', ascending=False) #los ordeno de mayor a menor
grouped_ordered_2.head(10) 
print(grouped_ordered_2.iloc[:5]['Country'])
```
